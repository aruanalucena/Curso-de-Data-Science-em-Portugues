# * Gradiente descedente, e regressão linear , não é unico que usa gradiente descedente principlamente redes neurais.
# Lógica 
**regressão tenta contruir uma reta que melhor descreva o comportamento dos dados**.Mais ajustada a relação com x e y 
* y=mx + b ( equação da reta)
m constante coeficiente angular
b coeficinete linear 
objetivo que melhor então um valor bem especifico de m e b.
* contruir um milhão de retas vou ver a melhor que melhor se adequada com os dados e vai pegar a melhor (grande despendio computacional só a sorte  mesmo)
* estratégia melhor 
y = mx+b
mudamos  **m** bem pequeno se ela esta melhor é um sinal de caminho e direção certa e a mesma coisa para **b** caminhando na direção certa se estiver ruim o caminho correto **é o inverso**.
muito **distante erro grande** vou variar bastante dar um passo maior e a partir que o **erro for menor começamos a caminhar de forma menor passos menores**, resumindo gradiente descedente
* Erro for grande , grandes passos e erro menor passos menores de pouco a pouco.

  * Sendo **k** o numero de partes em teste e k-1 para treino
k subdivisão =5-1teste = 4treino 
Vai fazer uma média dos k/k
     * Cross Validation mais poderoso um valor que vai chegar mais próximo da médias, um pouco mais veridico.evita viés.
    * Lado negtivo = Custo computacional alto data set muito alto, tem que avaliar, em geral entre 5 e 10 partes.

```
def  modelosregressaokfold(a, b):
        from sklearn.linear_model import LinearRegression
        from sklearn.linear_model import Ridge
        from sklearn.linear_model import Lasso
        from sklearn.linear_model import ElasticNet
        kfold = KFold(n_splits=10)
        x = a
        y = b
        reg = LinearRegression()
        ridge = Ridge()
        lasso = Lasso()
        elastic = ElasticNet()
        resul_reg = cross_val_score ( reg, x, y, cv =  kfold)
        resul_ridge = cross_val_score (reg, x, y, cv= kfold)
        resul_lasso = cross_val_score( reg, x, y, cv= kfold)
        resul_elastic =cross_val_score(reg, x, y, cv= kfold)
        dic_regmodels = {'Linear':resul_reg.mean(),'Ridge':resul_ridge.mean(),'Lasso':resul_lasso.mean(),'Elastic':resul_elastic.mean()}
        melhor_modelo =  max (dic_regmodels, key= dic_regmodels.get)
        print ('Regressão Linear :', resul_reg, 'Regressão Ridge:', resul_ridge, 'Reressão Lasso:',resul_lasso, 'Regressão Elastic:', resul_elastic)
        print ('O Melhor modelo foi :', melhor_modelo, 'com valor:', dic_regmodels[melhor_modelo])
```

```
modelosregressaokfold(x, y)
```

# Curva ROC  (Receiver Operating Characteristic) 
* **Taxa de verdadeiro positivo (True Positive Rate), que é dado por true positives / (true positives + false negatives)**. Essa taxa também é conhecida como sensibilidade, recordação ou probabilidade de detecção 
* **Taxa de falso positivo (False Positive Rate), que é dado por false positives / (false positives + true negatives)**. A taxa de falsos positivos também é conhecida como probabilidade de alarme falso e pode ser calculada como (1 — Specificity). A Specificity (especificidade) também é conhecida como true negative rate. Por exemplo, quantas pessoas sem uma doença (true negative) foi classificadas como sadias.

#  AUC (Area Under The Curve)
* Assim, para simplificar a curva ROC, foi criada a AUC. A AUC resume a curva ROC num único valor, calculando a “área sob a curva”.

 # ** INTERESSANTE** 
 * Ter uma curva ROC podemos comparar diferentes algoritimos ou mesmo algoritimos com parâmetros diferentes.
 * A area maior é melhor.

# Decision Trees Classifier ( ajustando parametros)

```
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
x = pd.DataFrame(iris.data, columns=[iris.feature_names])
y = pd.Series(iris.target)

import numpy as np
from sklearn.model_selection import  GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Definindo os valores que serão testados em DecisionTree
minimos_split  =  np.array ([2,3,4,5,6,8])
maximo_nivel  =  np.array ([3,4,5,6])
algoritimo = ['gini', 'entropy']
valores_grid = {'min_samples_split' : minimos_split, 'max_depth' : maximo_nivel, 'criterion' : algoritimo}

# Criação do modelo:
modelo =  DecisionTreeClassifier()

#Criando os Grids:
gridDecisionTree = GridSearchCV(estimator = modelo, param_grid = valores_grid, cv=5)
gridDecisionTree.fit(x,y)

#Imprimindo os melhores parametros:
print("Mínimo split: ", gridDecisionTree.best_estimator_.min_samples_split)
print("Máxima profundidade: ", gridDecisionTree.best_estimator_.max_depth)
print("Algoritimo escolhido : ", gridDecisionTree.best_estimator_.criterion )
print("Acuracia: ", gridDecisionTree.best_score_)

```

# Feature Selection : qui-quadrado(aplicação prática em Python)


```
from sklearn.feature_selection import  SelectKBest
from sklearn.feature_selection import  chi2

# Definindo variaveis preditoras e target
x = [[12, 2, 30] ,  [15, 11, 6] , [16, 8, 90], [5, 3, 20], [4, 14, 5] , [2, 5, 70]]
y = [1,1,1,0,0,0]

# Selecionando duas váriaveis com o maio chi-quadrado:
algoritimo =SelectKBest(score_func=chi2, k=2)
dados_das_melhores_preditoras = algoritimo.fit_transform(x,y)

# Resultados 
print('Scores : ', algoritimo.scores_)
print('Resultados da transformação:\n',  dados_das_melhores_preditoras)
```

```
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.feature_selection import  SelectKBest
from sklearn.feature_selection import chi2

# Carregando o dataset
iris = load_iris()

# Definindo as váriaveis preditoras e a váriavel target
x = pd.DataFrame(iris.data, columns=[iris.feature_names])
y = pd.Series(iris.target)

print(x.head())
```

```
# Selecionando  duas variaveis com o maior chi-quadrado:
algoritimo = SelectKBest(score_func=chi2, k=2)
dados_das_melhores_preditoras = algoritimo.fit_transform(x, y)

#  Resultados
print('Scores: ', algoritimo.scores_)
print('Resultado da transformação:\n' , dados_das_melhores_preditoras)
```
